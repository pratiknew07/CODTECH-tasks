```python
# Scatter plot for numerical variables
sns.scatterplot(x='age', y='fare', data=df)
plt.title('Age vs Fare')

plt.show()

# Box plot for categorical vs numerical variables
sns.boxplot(x='sex', y='age', data=df)
plt.title('Age Distribution by Gender')
plt.show()
```

```python
# Correlation matrix
corr_matrix = df.corr()

# Heatmap of the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Box plot to detect outliers in 'fare'
sns.boxplot(x=df['fare'])
plt.title('Box Plot of Fare')

plt.show()

# Removing outliers (example: fares above the 95th percentile)
q_high = df['fare'].quantile(0.95)
df = df[df['fare'] < q_high]
```

```python
# Standardize column names
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Correcting data entry errors (if any)
# Example: Replace 'male' with 'M' and 'female' with 'F'
df['sex'] = df['sex'].replace({'male': 'M', 'female': 'F'})
```

```python
df.head()
```

```python
# Creating new features
df['family_size'] = df['sibsp'] + df['parch'] + 1

# Log transformation to handle skewness
df['log_fare'] = np.log1p(df['fare'])

# Encoding categorical variables
df['sex'] = df['sex'].map({'male': 0, 'female': 1})
df = pd.get_dummies(df, columns=['embarked'], drop_first=True)
```

```python
df.head()
```

```python
sns.pairplot(df[['age', 'fare', 'family_size', 'survived']], hue='survived')
plt.show()

g = sns.FacetGrid(df, col='survived', row='sex', margin_titles=True)
g.map(plt.hist, 'age', bins=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame for illustration
data = {
    'date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
    'value': [10, 15, 7, 12, 9]
}
df = pd.DataFrame(data)

# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Set the 'date' column as the index
df.set_index('date', inplace=True)

# Plot the 'value' column over time
df['value'].plot()
plt.title('Value Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```

```python
# Slice data for a specific group
df_females = df[df['sex'] == 1]

# Analyze the subset
sns.countplot(x='class', hue='survived', data=df_females)
plt.title('Survival Rate by Class for Females')
plt.show()
```

```python
# Violin plot to show the distribution of age by survival status
sns.violinplot(x='survived', y='age', data=df, inner='quartile')
plt.title('Age Distribution by Survival Status')
plt.show()

# Swarm plot to show individual data points
sns.swarmplot(x='class', y='age', hue='survived', data=df)
plt.title('Age and Survival by Class')
plt.show()
```

```python
# Interaction between 'age' and 'fare' with survival status
sns.lmplot(x='age', y='fare', hue='survived', data=df, aspect=2, fit_reg=False)
plt.title('Age vs Fare by Survival Status')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df.select_dtypes(include=[np.number]))

# Apply PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)
df['pca1'] = pca_result[:, 0]
df['pca2'] = pca_result[:, 1]

# Plot PCA results
sns.scatterplot(x='pca1', y='pca2', hue='survived', data=df)
plt.title('PCA of Titanic Dataset')
plt.show()
```

```python
from scipy.stats import ttest_ind

# T-test for age difference between survived and not survived
t_stat, p_val = ttest_ind(df[df['survived'] == 1]['age'].dropna(), df[df['survived'] == 0]['age'].dropna())
print(f'T-test result: t-statistic={t_stat}, p-value={p_val}')
```
